```python
from nltk.tokenize import TweetTokenizer
from nltk.corpus import stopwords
import re
import enchant

d = enchant.Dict("en_US")
tt = TweetTokenizer()
stopwords = set(stopwords.words('english'))

def preprocess(lst):
    temp = []
    #list of tweets
    for tweet in lst:
        tweet = tweet.strip().split(" ")
        tweet = list(map(lambda x: x.lower(), tweet))
        #remove any word that does not contain any English alphabets
        tweet = list(map(lambda x: re.sub("^[^a-z\d]+$", "", x).strip(), tweet))
        tweet = list(map(lambda x: re.sub("^(http).*$", "", x).strip(), tweet))
        tweet = list(map(lambda x: re.sub("[^\w\s@#]", "", x).strip(), tweet))
        tweet = list(map(lambda x: re.sub("^[@].*$", "", x).strip(), tweet))
        tweet = list(map(lambda x: re.sub("\n", " ", x).strip(), tweet))
        #remove all '' in the list
        tweet = list(filter(lambda x: x != '', tweet))
        tweet = list(filter(lambda x: d.check(x), tweet))
        tweet = ' '.join(tweet)
        temp.append(tweet)
    return temp

def clean_text(
    string: str, 
    punctuations=r'''!()-[]{};:'"\,<>./?@#$%^&*_~''',
    stop_words=['the', 'a', 'and', 'is', 'be', 'will']) -> str:
    """
    A method to clean text 
    """
    # Cleaning the urls
    string = re.sub(r'https?://\S+|www\.\S+', '', string)

    # Cleaning the html elements
    string = re.sub(r'<.*?>', '', string)

    # Removing the punctuations
    for x in string.lower(): 
        if x in punctuations: 
            string = string.replace(x, "") 

    # Converting the text to lower
    string = string.lower()

    # Removing stop words
    string = ' '.join([word for word in string.split() if word not in stop_words])

    # Cleaning the whitespaces
    string = re.sub(r'\s+', ' ', string).strip()

    return string

def preprocess2(data):
    lst = []
    for tweet in data:
        temp = tweet.strip().split(" ")
        temp = list(map(lambda x : clean_text(x),temp))
        temp = ' '.join(temp)
        lst.append(temp)
    return lst
        
        
text_data = preprocess2(tra_data)
print(text_data[1])
text_dev_data = preprocess2(dev_data)
print(text_dev_data[0])

test_id = open("test.data.txt", 'r').readlines()
test_data = []
for i in range(len(test_id)):
    ids = test_id[i]
    ids = ids.strip().split(",")
    temp = []
    for ele in ids:
        temp.append(json.load(open("tweet-objects/tweet-objects/" + str(ele) + ".json")))
    temp2 = sorted(temp[1:] , key = lambda x : time.mktime(time.strptime(x['created_at'], "%a %b %d %H:%M:%S %z %Y")))
    temp3 = [temp[0]] + temp2
    temp3 = list(map(lambda x : x['text'],temp3)) 
    test_data.append(temp3)
    
text_test_data = []
for ele in test_data:
    temp = preprocess2(ele)
    text_test_data += temp
print(text_test_data[0])
```

```python
from keras.layers import GRU,Bidirectional, Masking,Dense,Dropout,LeakyReLU
import keras.backend as K
import tensorflow as tf
from keras.models import Sequential
from keras import regularizers

def get_f1(y_true, y_pred): #taken from old keras source code
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    recall = true_positives / (possible_positives + K.epsilon())
    f1_val = 2*(precision*recall)/(precision+recall+K.epsilon())
    return f1_val


model2 = Sequential(name="gru")
model2.add(Masking(0.,input_shape=(232, 768)))
lru = LeakyReLU(alpha=0.3)
model2.add(GRU(100,dropout=0.2,recurrent_dropout=0.2,activity_regularizer=regularizers.l2(3e-4),return_sequences=False))
#model2.add(Bidirectional(GRU(100,dropout=0.4,recurrent_dropout=0.4,return_sequences=True)))
#model2.add(GRU(10,return_sequences=False))
model2.add(Dense(10,activation='relu'))
#model2.add(Dropout(0.5))
model2.add(Dense(1,activation='sigmoid',activity_regularizer=regularizers.l2(3e-4)))
opt = tf.keras.optimizers.Adam(learning_rate = 0.01)
recall = tf.keras.metrics.Recall()
model2.compile(optimizer=opt,
              loss='binary_crossentropy',
              metrics=['accuracy',get_f1])
model2.summary()
```
